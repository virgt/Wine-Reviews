>For example, a review with the following description:
>
>"Very strong taste like apple and cinnamon"
>
>should have a `"length"` key with a value of 7.
>
>What's the length of the last review?


```python
def get_review_lenght(wine_reviews): 
    for review in wine_reviews: 
        review['lenght'] = len(review["description"].split())
    
    return wine_reviews
```

```python
# Adding description lenght to each review 
wine_reviews = get_review_lenght(wine_reviews)
```

## Question 5

>
>{country: average points of wines coming from that country}

>
>{"Spain": 9.25, "Italy": 9.5}

>Paste the full dictionary as the answer.

```python
average_point_by_country = {}
for country in set_countries:
    points = [int(review['points']) for review in wine_reviews if review['country'] == country]
    average_point_by_country[country] = sum(points)/len(points) #defini la key c'est [] = defini la value 

print (average_point_by_country)
```

## Question 6
>What's the country that produces the wines with the highest average rating?

```python
average_point_by_country = {}
for country in set_countries:
    points = [int(review['points']) for review in wine_reviews if review['country'] == country]
    average_point_by_country[country] = sum(points)/len(points) #defini la key c'est [] = defini la value 

average_point_by_country
```

```python
max_average = max(average_point_by_country, key=average_point_by_country.get)
print("the country that produces the wines with the highest average rating is", max_average)
```

### Question 7
>Update each wine's description by adding at the end of each description the following piece of text:
>
>"This is a {designation} from {country} that scored {points} points"
>
>What is the resulting description of the last review?

```python
def get_longer_description(wine_reviews): 
    for review in wine_reviews: 
        added_description = f'this is a {review["designation"]} from {review["country"]} that scored {review["points"]} points'
        if added_description in review['description']:
            pass 
        else: 
            review["description"]+= added_description
    return wine_reviews

#get_longer_description(wine_reviews)
```

### Question 8
>What's the proportion of wine tasters that have a Twitter account?

```python
all_twitter_account=[]

for review in wine_reviews: 
    if review['taster_twitter_handle'] != None: 
        all_twitter_account.append(review['taster_twitter_handle'])
```

```python
all__unique_tasters_with_twitter_account=set(all_twitter_account)
all__unique_tasters_with_twitter_account
```

```python
all_registred_account = []
for review in wine_reviews: 
    if review['taster_name']!= None: 
        all_registred_account.append(review['taster_name'])
    
all_unique_registred_account = set(all_registred_account)
all_unique_registred_account
```

```python
prop_taster_tweeter_account = (len(all__unique_tasters_with_twitter_account)/len(all_unique_registred_account))*100
print("There is ", prop_taster_tweeter_account, "% of wine tasters that have a Twitter account")
```
